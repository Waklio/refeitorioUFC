```python
#variável que controla os meses corretos.
month_control = 1
#dicionário em python vai fazer correspondncia do mês com a linha.
correspondent_month = {
  1: '01',
  16:'02',
  31:'03',
  46:'04',
  61:"05",
  76:'06',
  91:'07',
  106:'08',
  121:'09',
  136:'10',
  151:'11',
  166:'12',
}
#contandor do mês começa no 0
counter_month_year=0
#controle a aplicação
while True:
#começa em 0 incrementado a cada mês a cada interação 
  counter_month_year += 1
#laço for controla a execução de cada mês
  for each_columns in fulldata2.columns:# fultdata2 retorna uma lista de todas colunas

# pega cada coluna exemplo almoço depois unnamend e converte pra string
    each_columns = str(each_columns)
#se ele for igual almoço sobral n precisamos
    if each_columns == 'ALMOÇO SOBRAL':
      continue 
    
    
#linha dos dias corretos recebe mês control soma mais 2 para aparecer a linha correta
    correct_days_line = month_control + 2
#mês verdadeiro vai receber o dicionario acima e retorna o valor ao correpondente (caso n encontre retorna none)
    real_month = correspondent_month.get(month_control)
#controle da linha  equivalente ao total e pra acertar os linha certa basta somar
    control_quantaty_line = correct_days_line + 11
# tratar erros através dessa estrutra(assim que acabar os dias )
    try:

     day_month = str(int(fulldata2[each_columns][correct_days_line]))

    except Exception as error:
      break #sair de laço for
#dias corretos, mês real, e dia do mês e joga no  dia atual
#      
      
    current_date = '2019/'+ real_month +'/'+ re.findall("\d+", day_month)[0] #re biblioteca pra pegar expressões regulares
    quantidade = fulldata2[each_columns][control_quantaty_line]
    print(f'{current_date} | {quantidade}')

#atualizar o controle do mês ou seja soma 1 posi
  month_control+= 15
  if counter_month_year >= 12:#teste pra saber se foi tratado todos os meses
    break #sair do laço while

```
