```python
#variÃ¡vel que controla os meses corretos.
month_control = 1
#dicionÃ¡rio em python vai fazer correspondncia do mÃªs com a linha.
correspondent_month = {
  1: '01',
  16:'02',
  31:'03',
  46:'04',
  61:"05",
  76:'06',
  91:'07',
  106:'08',
  121:'09',
  136:'10',
  151:'11',
  166:'12',
}
#contandor do mÃªs comeÃ§a no 0
counter_month_year=0
#controle a aplicaÃ§Ã£o
while True:
#comeÃ§a em 0 incrementado a cada mÃªs a cada interaÃ§Ã£o 
  counter_month_year += 1
#laÃ§o for controla a execuÃ§Ã£o de cada mÃªs
  for each_columns in fulldata2.columns:# fultdata2 retorna uma lista de todas colunas

# pega cada coluna exemplo almoÃ§o depois unnamend e converte pra string
    each_columns = str(each_columns)
#se ele for igual almoÃ§o sobral n precisamos
    if each_columns == 'ALMOÃ‡O SOBRAL':
      continue 
    
    
#linha dos dias corretos recebe mÃªs control soma mais 2 para aparecer a linha correta
    correct_days_line = month_control + 2
#mÃªs verdadeiro vai receber o dicionario acima e retorna o valor ao correpondente (caso n encontre retorna none)
    real_month = correspondent_month.get(month_control)
#controle da linha  equivalente ao total e pra acertar os linha certa basta somar
    control_quantaty_line = correct_days_line + 11
# tratar erros atravÃ©s dessa estrutra(assim que acabar os dias )
    try:

     day_month = str(int(fulldata2[each_columns][correct_days_line]))

    except Exception as error:
      break #sair de laÃ§o for
#dias corretos, mÃªs real, e dia do mÃªs e joga no  dia atual
#      
      
    current_date = '2019/'+ real_month +'/'+ re.findall("\d+", day_month)[0] #re biblioteca pra pegar expressÃµes regulares
    quantidade = fulldata2[each_columns][control_quantaty_line]
    print(f'{current_date} | {quantidade}')

#atualizar o controle do mÃªs ou seja soma 1 posi
  month_control+= 15
  if counter_month_year >= 12:#teste pra saber se foi tratado todos os meses
    break #sair do laÃ§o while

```
